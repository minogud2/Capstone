```python
def check_country(foo, gdf):
    count = 0
    while count < len(gdf):
        test = foo['geometry'].within(gdf['geometry'].iloc[count])
        if test == True:
            new_region = gdf['ADM0_PCODE'].iloc[count]
            return new_region
        else:
            count +=1
```

```python
def check_region(foo, gdf):
    count = 0
    while count < len(gdf):
        test = foo['geometry'].within(gdf['geometry'].iloc[count])
        if test == True:
            new_region = gdf['ADM1_PCODE'].iloc[count]
            return new_region
        else:
            count +=1
```

```python
def check_zone(foo, gdf):
    count = 0
    while count < len(gdf):
        test = foo['geometry'].within(gdf['geometry'].iloc[count])
        if test == True:
            new_zone = gdf['ADM2_PCODE'].iloc[count]
            return new_zone
        else:
            count +=1
```

```python
def check_woreda(foo, gdf):
    count = 0
    while count < len(gdf):
        test = foo['geometry'].within(gdf['geometry'].iloc[count])
        if test == True:
            new_woreda = gdf['ADM3_PCODE'].iloc[count]
            return new_woreda
        else:
            count +=1
```

```python
import json
```

```python
gdf_countries = json.load(open('geo/eth_admin0.json', 'r'))
```

```python
gdf_regions = json.load(open('geo/eth_admin1.json', 'r'))
gdf_zones = json.load(open('geo/eth_admin2.json', 'r'))
gdf_woredas = json.load(open('geo/eth_admin3.json', 'r'))
```

```python
country_id_map = {}
for feature in gdf_countries['features']:
    feature['id'] = feature['properties']['ADM0_PCODE']
    country_id_map[feature['properties']['ADM0_EN']] = feature['id']
```

```python
# plotly needs you to specify a property called ID in order to render graph. 
# Need to create this id with UNIQUE identifier -i.e. p codes. 
region_id_map = {}
for feature in gdf_regions['features']:
    feature['id'] = feature['properties']['ADM1_PCODE']
    region_id_map[feature['properties']['ADM1_EN']] = feature['id']

zone_id_map = {}
for feature in gdf_zones['features']:
    feature['id'] = feature['properties']['ADM2_PCODE']
    zone_id_map[feature['properties']['ADM2_EN']] = feature['id']

woreda_id_map = {}
for feature in gdf_woredas['features']:
    feature['id'] = feature['properties']['ADM3_PCODE']
    woreda_id_map[feature['properties']['ADM3_EN']] = feature['id']
```

```python
with open("eth_admin1v2.json", "w") as outfile: 
    json.dump(gdf_regions, outfile) 

with open("eth_admin2v2.json", "w") as outfile: 
    json.dump(gdf_zones, outfile) 
    
with open("eth_admin3v2.json", "w") as outfile: 
    json.dump(gdf_woredas, outfile) 
```

```python
with open("eth_admin0v2.json", "w") as outfile: 
    json.dump(gdf_countries, outfile) 
```
